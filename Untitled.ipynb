```python
for i in range(len(data_df)):
    raw_data = data_df.iloc[i]['final_price_list']
    if type(raw_data) == float:
        continue
    result = match.search(raw_data)
    if result:
        price = int(result.group(1).replace(',',''))
    else:
        price = -1
    data_df.set_value(i, 'final_price_list', price)
```
